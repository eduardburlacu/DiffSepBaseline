# Import Required Libraries
Import the necessary libraries, including json, matplotlib, and numpy.

```python
# Import the necessary libraries
import json
import matplotlib.pyplot as plt
import numpy as np
```

# Load JSON Data
Load the JSON data from the specified file path.

```python
# Load JSON Data

# Define the file path
file_path = 'diffusion-separation/results/diffsep_checkpoint_N-30_snr-0.5_corrstep-1_denoise-True_schedule-None/libri-clean.json'

# Load the JSON data
with open(file_path, 'r') as file:
    data = json.load(file)

# Display the loaded data (optional, for verification)
print(data)
```

# Extract Metrics
Extract the metrics (si_sdr, si_sir, si_sar, pesq, stoi) from the JSON data.

```python
# Extract Metrics

# Initialize lists to store the metrics
si_sdr_1 = si_sdr_2 = []
si_sir_1= si_sir_2 = []
si_sar_1 = si_sar_2 = []
pesq_1 = pesq_2 = []
stoi_1 = stoi_2 = []

# Iterate through the JSON data and extract the metrics
for key, value in data.items():
    si_sdr_1.append(value['si_sdr'][0][0])
    si_sir_1.append(value['si_sir'][0][0])
    si_sar_1.append(value['si_sar'][0][0])
    pesq_1.append(value['pesq'][0])
    stoi_1.append(value['stoi'][0])
    si_sdr_2.append(value['si_sdr'][0][1])
    si_sir_2.append(value['si_sir'][0][1])
    si_sar_2.append(value['si_sar'][0][1])
    pesq_2.append(value['pesq'][1])
    stoi_2.append(value['stoi'][1])

# Convert lists to numpy arrays for easier manipulation
si_sdr_1 = np.array(si_sdr_1)
si_sir_1 = np.array(si_sir_1)
si_sar_1 = np.array(si_sar_1)
pesq_1 = np.array(pesq_1)
stoi_1 = np.array(stoi_1)
si_sdr_2 = np.array(si_sdr_2)
si_sir_2 = np.array(si_sir_2)
si_sar_2 = np.array(si_sar_2)
pesq_2 = np.array(pesq_2)
stoi_2 = np.array(stoi_2)

# Display the extracted metrics (optional, for verification)
print("_______Speaker 1_______")
print("SI-SDR:", si_sdr_1.mean())
print("SI-SIR :", si_sir_1.mean())
print("SI-SAR:", si_sar_1.mean())
print("PESQ:", pesq_1.mean())
print("STOI:", stoi_1.mean())
print("_______Speaker 2_______")
print("SI-SDR:", si_sdr_2.mean())
print("SI-SIR:", si_sir_2.mean())
print("SI-SAR:", si_sar_2.mean())
print("PESQ:", pesq_2.mean())
print("STOI:", stoi_2.mean())
```

# Plot Histograms
